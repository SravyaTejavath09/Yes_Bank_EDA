## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'], format='%b-%d')

# Extract year and month from 'Date'
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Create a new column for day of the week
df['DayOfWeek'] = df['Date'].dt.dayofweek  # Monday=0, Tuesday=1, ..., Sunday=6

# Create a new column for quarter of the year
df['Quarter'] = df['Date'].dt.quarter

# Create a column for the month name
df['MonthName'] = df['Date'].dt.month_name()

# Create a column for the day name
df['DayName'] = df['Date'].dt.day_name()

# Create a new column for week of the year (workaround for older Pandas versions)
df['WeekOfYear'] = df['Date'].dt.strftime('%U').astype(int)
````

### What all manipulations have you done and insights you found?

### Data Manipulations:


*   Converted 'Date' to Datetime: The 'Date' column, originally a string, was converted to datetime objects using pd.to_datetime with the appropriate format (%b-%d). This enables time-based analysis and calculations.
*   Extracted Time Components: New columns were created to extract year, month, day of the week, quarter, month name, day name, and week of the year from the 'Date' column. These provide granular time-based features for analysis.


### Potential Insights:
*   Trends and Seasonality: Analyzing stock prices over different time periods (years, months, quarters, weeks) can reveal trends and seasonality patterns.
